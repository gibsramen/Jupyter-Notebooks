<h1>Introduction</h1>

Analysis of Pokemon types in Smogon 6v6.

<h1>Data Wrangling</h1>

```r
# my favorite libraries
library(ggplot2)
library(reshape2)
library(repr)
```

Importing the data is always useful.

```r
d <- read.csv('smogon.csv', header = TRUE, stringsAsFactors = FALSE)
colnames(d)[5:11] <- c('BST', 'HP', 'Atk', 'Def', 'SpA', 'SpD', 'Spd')

head(d)
```

Since all the 18 types appear as primary types at least once, we can save some typing by just using the unique values as a list of all types.

```r
all_types <- unique(d$Type.1)
all_types
```

Now we can start looking at the stats by typing. We'll start by isolating all the stats as a list and then melting the data.

```r
all_stats = names(d)[5:11]

stats <- melt(d,
              id = c('Name', 'Type.1', 'Type.2', 'Tier', 'Legendary', 'Mega', 'Generation'),
              measure.vars = all_stats)

colnames(stats)[8:9] <- c('Stat', 'Value')
bst_stats <- subset(stats, Stat=='BST')
i_stats <- subset(stats, Stat!='BST')

head(stats)
```

Hmmm... the Pokemon with two types are problematic. What can I do about that? We could use only the primary type but that's losing a lot of data. Let's take every double type Pokemon on the list and have them appear twice per stat, once for each type.

```r
dual <- stats[which(stats$Type.2!=''),]
dual$Type.1 <- dual$Type.2
dual$Type.2 <- NULL
head(dual)
```

Awesome. Now we can add the original list without the secondary types.

```r
stats.df <- rbind(stats[-3], dual)
colnames(stats.df) <- c('Name', 'Type', 'Tier', 'Legendary', 
                        'Mega', 'Generation', 'Stat', 'Value')
```

Finally, we can change the levels of 'Legendary' and 'Mega' for clarity.

```r
stats.df$Mega <- as.factor(stats.df$Mega)
levels(stats.df$Mega) <- c('Non-Mega', 'Mega')

stats.df$Legendary <- as.factor(stats.df$Legendary)
levels(stats.df$Legendary) <- c('Non-Legendary', 'Legendary')
```

Because BST and individual stats are so different, I am going to separate them into two data frames.

```r
ind_stats.df = subset(stats.df, Stat!='BST')
total_stats.df = subset(stats.df, Stat=='BST')

head(stats.df)
```

<h1>Typing</h1>

Let's look at the median value for each stat for each type.

```r
type.df <- dcast(stats.df, Type ~ Stat, median, fill=NaN, value.var = 'Value')
head(type.df)
```

```r
ind_type.df <- type.df[-2]

m.type.df <- melt(type.df, id='Type')
colnames(m.type.df) <- c('Type', 'Stat', 'Value')
head(m.type.df)
```

Let's look at the plot.

```r
ggplot(data=subset(m.type.df, Stat!='BST'), aes(x=Type, y=Stat)) +
    geom_tile(aes(fill=Value), color='black') +
    coord_flip() +
    scale_fill_gradient(low='white', high='purple4', limits=c(50,120)) +
    ggtitle('Median Stats by Type')
```

We can get a lot of data by looking at this. From a cursory inspection we can see that Dragon and Fighting types dominate the Attack stat, while Rock and Steel types dominate the Defense stat. Just as expected.

Median BST we can see on a scatter plot.

```r
x <- subset(m.type.df, Stat=='BST')
ggplot(data=x, aes(x=reorder(Type, Value), y=Value)) +
    geom_point(stat="identity") +
    coord_flip() +
    ggtitle('Median BST by Type')
```

<h1>Mega Evolutions and Legendaries</h1>

Mega Evolution and Legendary Pokemon are a huge part of the Smogon metagame. Let's look at the individual stat distribution of these groups.

```r
ggplot(data=i_stats, aes(x=Stat, y=Value)) +
    geom_boxplot(aes(fill=Stat)) +
    facet_grid(Legendary ~ Mega) +
    ggtitle('Stat Distribution') +
    guides(fill=FALSE) +
    scale_y_continuous(limits=c(0,255), breaks=seq(0,250,50))
```

Now for BST.

```r
ggplot(data=bst_stats, aes(x=Stat, y=Value, fill=Stat)) +
    geom_boxplot() +
    facet_grid(Legendary ~ Mega) +
    ggtitle('BST Distribution') +
    guides(fill=FALSE) +
    scale_y_continuous(limits=c(200,800), breaks=seq(200,800,100))
```

<h1>Generation</h1>

Generation does not directly affect a Pokemon's viability but let's poke around and see what we can find.

```r
ggplot(data=i_stats, aes(x=Stat, y=Value)) +
    geom_boxplot(aes(fill=Stat)) +
    facet_wrap(~ Generation) +
    scale_y_continuous(limits=c(0,255), breaks=seq(0,250,50)) +
    guides(fill=FALSE) +
    scale_x_discrete(limits=c('HP', 'Atk', 'Def', 'SpA', 'SpD', 'Spd')) +
    ggtitle('Stat Distribution by Generation')
```

This is kind of hard to read. Maybe a heatmap showing the medians would be better.

```r
gen.df <- dcast(stats, Generation ~ Stat, median, fill=NaN, value.var='Value')
m.gen.df <- melt(gen.df, id='Generation')
colnames(m.gen.df) <- c('Generation', 'Stat', 'Value')
head(m.gen.df)
```

```r
ggplot(data=subset(m.gen.df, Stat!='BST'), aes(x=Generation, y=Stat)) +
    geom_tile(aes(fill=Value), color='black') +
    scale_fill_gradient(low='white', high='darkred') +
    scale_x_continuous(breaks=seq(1,6,1)) +
    coord_flip() +
    geom_text(aes(label=Value)) +
    ggtitle('Median Stats by Generation')
```

Now for BST (again).

```r
ggplot(data=subset(m.gen.df, Stat=='BST'), aes(x=Generation, y=Value)) +
    geom_bar(stat='identity', color='black', fill='orange3') +
    coord_cartesian(ylim=c(470, 520)) +
    scale_x_continuous(breaks=seq(1,6,1)) +
    ggtitle('Median BST By Generation')
```

Looks like overall, Gen II had the lowest BST Pokemon and Gen IV had the highest. However, it is important to note that the only Pokemon in this dataset are those in Smogon 6v6.

<h1>Tier</h1>

Tier is likely the most important factor when considering a Pokemon's stats (or really vice-versa).

What I'm going to do for this analysis is a bit of data consolidation: that is, I'm going to eliminate the 'middle' tiers. This means that OU will consist of all Pokemon in OU and BL, UU will consist of all Pokemon in UU and BL2, etc. I am going to include AG with Ubers seeing as Mega Rayquaza is the only denizen of that tier and his stats are more in line with Ubers than anything else.

```r
smog_stats.df <- stats

head(smog_stats.df)
```

```r
smog_stats.df$Tier[which(smog_stats.df$Tier=='BL')] <- 'OU'
smog_stats.df$Tier[which(smog_stats.df$Tier=='BL2')] <- 'UU'
smog_stats.df$Tier[which(smog_stats.df$Tier=='BL3')] <- 'RU'
smog_stats.df$Tier[which(smog_stats.df$Tier=='BL4')] <- 'NU'
smog_stats.df$Tier[which(smog_stats.df$Tier=='AG')] <- 'Uber'
```

```r
c.smog_stats <- dcast(smog_stats.df, Tier ~ Stat, median, fill = NaN, value.var='Value')
c.smog_stats
```

```r
m.smog_stats <- melt(c.smog_stats, id='Tier')
colnames(m.smog_stats) <- c('Tier', 'Stat', 'Value')
head(m.smog_stats)
```

```r
ggplot(data=subset(m.smog_stats, Stat!='BST'), aes(x=Tier, y=Stat, fill=Value)) +
    geom_tile(color='black') +
    scale_fill_gradient(low='white', high='green4') +
    scale_x_discrete(limits=c('Uber', 'OU', 'UU', 'NU', 'RU', 'PU')) +
    coord_flip() +
    geom_text(aes(label=Value)) +
    ggtitle('Median Stats by Tier')
```

```r
ggplot(data=subset(m.smog_stats, Stat=='BST'), aes(x=Tier, y=Value)) +
    geom_bar(stat='identity', color='black', fill='green3') +
    coord_cartesian(ylim=c(400, 700)) +
    scale_x_discrete(limits=c('Uber', 'OU', 'UU', 'RU', 'NU', 'PU')) +
    scale_y_continuous(breaks=seq(400,700,50)) +
    ggtitle('Median BST by Tier')
```
