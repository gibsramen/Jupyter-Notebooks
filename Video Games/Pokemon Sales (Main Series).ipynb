```r
library(ggplot2)
library(repr)
library(reshape2)
```

```r
sales <- read.csv('Video_Games_Sales.csv', header=T, stringsAsFactors=F)
head(sales)
```

```r
pk.sales <- sales[grepl('Pokemon.*\\/', sales$Name) & 
                  sales$Global_Sales >= 1 &
                  !grepl('Mystery', sales$Name),]
drops = c('Genre', 'Publisher')
pk.sales <- pk.sales[, !names(pk.sales) %in% drops]
pk.sales <- pk.sales[, -9:-14]
```

```r
pk.sales
```

```r
m.pk.sales <- melt(pk.sales,
                   id=c('Name', 'Platform', 'Year_of_Release'))
```

```r
m.pk.sales$Year_of_Release <- as.numeric(m.pk.sales$Year_of_Release)
head(m.pk.sales)
```

```r
ggplot(data=m.pk.sales, aes(x=Year_of_Release, y=value, group=variable)) +
    geom_line(aes(colour=variable), size=1) +
    scale_x_continuous(lim=c(1996, 2016), breaks=seq(1996, 2016, 2))
```
